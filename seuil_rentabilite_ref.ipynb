## üéØ Partie 2 : Concepts Cl√©s - Avant les Calculs

### üí° Comprendre les Diff√©rents Types de Co√ªts

#### 1Ô∏è‚É£ **Charges Variables (CV)** - Changent avec la production
Ce sont les co√ªts qui **augmentent quand vous vendez plus** et **diminuent quand vous vendez moins**.

**Exemples :**
- Mati√®re premi√®re pour fabriquer un produit
- Emballage pour chaque produit
- Commissions de vente (% du chiffre d'affaires)
- Frais de livraison par colis

**Important** : Si vous ne vendez RIEN, vous ne payez PAS ces co√ªts !

#### 2Ô∏è‚É£ **Charges Fixes (CF)** - Restent les m√™mes chaque mois
Ce sont les co√ªts **ind√©pendants du chiffre d'affaires**. Vous les payez m√™me si vous ne vendez rien !

**Exemples :**
- Loyer du local (2 000 TND/mois = m√™me co√ªt en janvier et d√©cembre)
- Salaire des employ√©s permanents
- Assurances
- √âlectricit√©/internet (partie fixe)
- Abonnements logiciels

#### 3Ô∏è‚É£ **Amortissement** - Co√ªt des investissements √† long terme
Quand vous achetez un √©quipement co√ªteux (ex: 10 000 TND), vous ne mettez pas tout en charge cette ann√©e-l√†.
Vous **r√©partissez le co√ªt** sur sa dur√©e de vie.

**Exemples :**
- Machine achet√©e 10 000 TND, dur√©e 5 ans ‚Üí 2 000 TND/an d'amortissement
- Ordinateur 3 000 TND, dur√©e 3 ans ‚Üí 1 000 TND/an d'amortissement

#### 4Ô∏è‚É£ **D√©penses Exceptionnelles** - Grandes d√©penses ponctuelles
Ce sont des co√ªts extraordinaires qui arrivent rarement et co√ªtent beaucoup (> 2500 TND).

**Exemples :**
- R√©novation du local : 15 000 TND (une fois tous les 10 ans)
- Achat de nouvel √©quipement : 8 000 TND (ponctuel)
- Frais l√©gaux exceptionnels : 2 000 TND (rare)

## üî¢ Partie 3 : Formules et D√©finitions

### Le Seuil de Rentabilit√© (SR) en 3 √©tapes :

**√âtape 1** : Calculer la Marge sur Co√ªts Variables
$$\text{MCV} = \text{Chiffre d'Affaires} - \text{Charges Variables}$$

**√âtape 2** : Calculer le Taux de Marge
$$\text{Taux MCV} = \frac{\text{MCV}}{\text{Chiffre d'Affaires}}$$

**√âtape 3** : Trouver le SR (le CA minimum pour couvrir les co√ªts fixes + amortissement)
$$\text{Seuil de Rentabilit√©} = \frac{\text{Co√ªts Fixes + Amortissement}}{\text{Taux MCV}}$$

### üìä Cas d'√©tude : Startup TECHPRO en Tunisie

L'entreprise TECHPRO vend des applications mobiles en Afrique du Nord.

**Donn√©es R√âALISTES pour la Tunisie :**
- Chiffre d'affaires (CA) : 180 000 TND/an (environ 15 000 TND/mois)
- Charges variables : 63 000 TND/an (co√ªts de serveurs, support client, commissions)
- Charges fixes : 72 000 TND/an (salaires d√©veloppeur, loyer bureau, internet)
- Amortissement des investissements : 12 000 TND/an (ordinateurs, logiciels)
- D√©penses exceptionnelles : 0 TND (cette ann√©e-ci)

```python
# Cr√©ation d'un dictionnaire avec les donn√©es
donnees = {
    'CA': 180_000,      # Chiffre d'affaires annuel
    'CV': 63_000,       # Charges variables
    'CF': 72_000,       # Charges fixes
    'Depreciation': 12_000,  # Amortissement
    'Depenses_exceptionnelles': 0  # D√©penses ponctuelles
}

print("üìä Donn√©es de l'entreprise TECHPRO (En TND) :")
print(f"Chiffre d'affaires : {donnees['CA']:,.0f} TND")
print(f"Charges variables  : {donnees['CV']:,.0f} TND")
print(f"Charges fixes      : {donnees['CF']:,.0f} TND")
print(f"Amortissement      : {donnees['Depreciation']:,.0f} TND")
print(f"D√©penses exceptionnelles : {donnees['Depenses_exceptionnelles']:,.0f} TND")
```

## üßÆ Partie 3 : Cr√©ation d'une Classe Python pour le Seuil de Rentabilit√©

Nous allons cr√©er une **classe** qui contient toutes les m√©thodes de calcul.
C'est une excellente pratique de programmation orient√©e objet !

```python
class SeuilRentabilite:
    """
    Classe pour calculer le seuil de rentabilit√© avec tous les indicateurs.
    
    Inclut:
    - Charges variables (changent avec les ventes)
    - Charges fixes (m√™me chaque mois)
    - Amortissement (√©talement des investissements)
    - D√©penses exceptionnelles (ponctuelles)
    
    Attributes:
        ca (float): Chiffre d'affaires annuel
        cv (float): Charges variables annuelles
        cf (float): Charges fixes annuelles
        depreciation (float): Amortissement annuel
        depenses_exceptionnelles (float): D√©penses ponctuelles
    """
    
    def __init__(self, ca, cv, cf, depreciation=0, depenses_exceptionnelles=0):
        """
        Initialise l'objet avec les donn√©es financi√®res.
        
        Args:
            ca (float): Chiffre d'affaires
            cv (float): Charges variables
            cf (float): Charges fixes
            depreciation (float): Amortissement annuel
            depenses_exceptionnelles (float): D√©penses exceptionnelles
        """
        self.ca = ca
        self.cv = cv
        self.cf = cf
        self.depreciation = depreciation
        self.depenses_exceptionnelles = depenses_exceptionnelles
    
    def marge_cv(self):
        """Calcule la Marge sur Co√ªt Variable (CA - CV)."""
        return self.ca - self.cv
    
    def taux_mcv(self):
        """Calcule le Taux de Marge sur Co√ªt Variable (%)."""
        return self.marge_cv() / self.ca
    
    def taux_cv(self):
        """Calcule le Taux de Charges Variables (%)."""
        return self.cv / self.ca
    
    def charges_fixes_totales(self):
        """Calcule les charges fixes totales (CF + Amortissement)."""
        return self.cf + self.depreciation
    
    def resultat_avant_depenses_exceptionnelles(self):
        """Calcule le r√©sultat AVANT les d√©penses exceptionnelles."""
        return self.marge_cv() - self.charges_fixes_totales()
    
    def resultat_final(self):
        """Calcule le r√©sultat NET (apr√®s d√©penses exceptionnelles)."""
        return self.resultat_avant_depenses_exceptionnelles() - self.depenses_exceptionnelles
    
    def seuil_rentabilite(self):
        """
        Calcule le Seuil de Rentabilit√© (CA minimum pour couvrir CF + Amortissement).
        C'est le CA n√©cessaire pour avoir un r√©sultat = 0.
        """
        return self.charges_fixes_totales() / self.taux_mcv()
    
    def point_mort(self, jours_annee=365):
        """Calcule le Point Mort (nombre de jours pour atteindre le SR)."""
        return (self.seuil_rentabilite() / self.ca) * jours_annee
    
    def marge_securite(self):
        """Calcule la Marge de S√©curit√© (CA - SR)."""
        return self.ca - self.seuil_rentabilite()
    
    def indice_securite(self):
        """Calcule l'Indice de S√©curit√© (marge de s√©curit√© en %)."""
        return self.marge_securite() / self.ca
    
    def indice_prelevement(self):
        """Calcule l'Indice de Pr√©l√®vement (charges fixes totales / CA)."""
        return self.charges_fixes_totales() / self.ca
    
    def levier_operationnel(self):
        """
        Calcule le Levier Op√©rationnel.
        Mesure la sensibilit√© du r√©sultat aux variations du CA.
        """
        resultat = self.resultat_avant_depenses_exceptionnelles()
        if resultat == 0:
            return float('inf')  # Infini au seuil
        return self.marge_cv() / resultat
    
    def ca_pour_resultat(self, resultat_souhaite):
        """Calcule le CA n√©cessaire pour atteindre un r√©sultat donn√©."""
        mcv_necessaire = resultat_souhaite + self.charges_fixes_totales()
        return mcv_necessaire / self.taux_mcv()
    
    def rapport_complet(self):
        """G√©n√®re un rapport complet avec tous les indicateurs."""
        return {
            'Chiffre d\'affaires': self.ca,
            'Charges variables': self.cv,
            'Marge sur CV': self.marge_cv(),
            'Taux MCV (%)': self.taux_mcv() * 100,
            'Charges fixes': self.cf,
            'Amortissement': self.depreciation,
            'Charges fixes totales': self.charges_fixes_totales(),
            'D√©penses exceptionnelles': self.depenses_exceptionnelles,
            'R√©sultat NET': self.resultat_final(),
            'Seuil de rentabilit√©': self.seuil_rentabilite(),
            'Point mort (jours)': self.point_mort(),
            'Levier op√©rationnel': self.levier_operationnel()
        }

print("‚úÖ Classe SeuilRentabilite cr√©√©e avec succ√®s !")
```

## üí° Partie 4 : Utilisation de la Classe

Cr√©ons un objet avec les donn√©es de l'entreprise TECHNO.

```python
# Cr√©ation d'une instance de la classe
entreprise = SeuilRentabilite(
    ca=donnees['CA'],
    cv=donnees['CV'],
    cf=donnees['CF'],
    depreciation=donnees['Depreciation'],
    depenses_exceptionnelles=donnees['Depenses_exceptionnelles']
)

# Affichage du rapport complet
rapport = entreprise.rapport_complet()

# Conversion en DataFrame pour un affichage √©l√©gant
df_rapport = pd.DataFrame(list(rapport.items()), columns=['Indicateur', 'Valeur'])
print("\nüìà RAPPORT COMPLET - ENTREPRISE TECHPRO (En TND)\n")
print(df_rapport.to_string(index=False))
```

## üìâ Exemple 2 : Graphique du Seuil de Rentabilit√©

```python
# G√©n√©ration de donn√©es pour le graphique
ca_range = np.linspace(0, 250_000, 100)
charges_fixes_total = np.full_like(ca_range, entreprise.charges_fixes_totales())
charges_variables = ca_range * entreprise.taux_cv()
charges_totales = charges_fixes_total + charges_variables
mcv = ca_range * entreprise.taux_mcv()

# Cr√©ation du graphique
fig, ax = plt.subplots(figsize=(14, 8))

# Trac√© des courbes
ax.plot(ca_range, mcv, label='Marge sur Co√ªt Variable', 
        linewidth=3, color='#06A77D', linestyle='-')
ax.plot(ca_range, charges_fixes_total, label='Charges Fixes + Amortissement', 
        linewidth=3, color='#C73E1D', linestyle='--')
ax.plot(ca_range, charges_totales, label='Charges Totales (Fixes + Variables + Amort)', 
        linewidth=2, color='#A23B72', linestyle='-.')
ax.plot(ca_range, ca_range, label='Chiffre d\'affaires', 
        linewidth=2, color='#2E86AB', linestyle=':')

# Marquage du seuil de rentabilit√©
sr = entreprise.seuil_rentabilite()
ax.axvline(x=sr, color='red', linestyle='--', linewidth=2, alpha=0.7)
ax.axhline(y=sr, color='red', linestyle='--', linewidth=2, alpha=0.7)
ax.plot(sr, sr, 'ro', markersize=12, label='Seuil de Rentabilit√©')

# Annotation du seuil
ax.annotate(f'SR = {sr:,.0f} TND',
            xy=(sr, sr), xytext=(sr + 20_000, sr - 15_000),
            fontsize=12, fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.7),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.3', lw=2))

# Zone de profit et de perte
ax.fill_between(ca_range, 0, ca_range, where=(ca_range >= sr), 
                alpha=0.2, color='green', label='Zone de Profit')
ax.fill_between(ca_range, 0, ca_range, where=(ca_range < sr), 
                alpha=0.2, color='red', label='Zone de Perte')

# Configuration des axes
ax.set_xlabel('Chiffre d\'affaires (TND)', fontsize=14, fontweight='bold')
ax.set_ylabel('Montant (TND)', fontsize=14, fontweight='bold')
ax.set_title('Graphique du Seuil de Rentabilit√© - TECHPRO', 
             fontsize=16, fontweight='bold', pad=20)
ax.legend(loc='upper left', fontsize=10, framealpha=0.9)
ax.grid(True, alpha=0.3)
ax.set_xlim(0, 250_000)
ax.set_ylim(0, 250_000)

# Format des nombres sur les axes
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000:.0f}k'))
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000:.0f}k'))

plt.tight_layout()
plt.show()

print(f"\nüéØ Le seuil de rentabilit√© est {sr:,.0f} TND")
print(f"üìÖ Soit au jour {entreprise.point_mort():.0f} de l'ann√©e")
```

## üìù R√âSUM√â - Les 6 Concepts Fondamentaux

### ‚ú® Ce que vous devez retenir :

Ce notebook vous pr√©sente les 6 concepts essentiels pour comprendre la rentabilit√© d'un projet :

#### 1Ô∏è‚É£ **Charges Variables (CV)** 
- Co√ªts qui **changent avec les ventes**
- Exemples : mati√®re premi√®re, emballage, frais de livraison
- **Si vous ne vendez rien, vous ne payez pas ces co√ªts**

#### 2Ô∏è‚É£ **Charges Fixes (CF)**
- Co√ªts **identiques chaque mois**, ind√©pendants des ventes
- Exemples : loyer, salaires, assurances
- **Vous les payez m√™me si vous ne vendez rien**

#### 3Ô∏è‚É£ **Amortissement**
- R√©partition des **investissements co√ªteux** sur plusieurs ann√©es
- Exemples : machine (5 ans), ordinateur (3 ans)
- **√âtale l'impact financier** dans le temps

#### 4Ô∏è‚É£ **D√©penses Exceptionnelles**
- **Grandes d√©penses ponctuelles** (> 2500 TND)
- Exemples : r√©novation du local, nouvel √©quipement
- **Arrivent rarement** mais co√ªtent beaucoup

#### 5Ô∏è‚É£ **Marge sur Co√ªts Variables (MCV)**
- C'est l'argent disponible apr√®s payer les CV
- **Formule simple** : MCV = CA - CV
- **Vous l'utilisez pour** payer les CF et faire du profit

#### 6Ô∏è‚É£ **Seuil de Rentabilit√© (SR)**
- Le CA **minimum** pour ne pas perdre d'argent
- **Formule** : SR = (CF + Amortissement) / Taux MCV
- **Au SR** : B√©n√©fice = 0 (point d'√©quilibre)

#### 7Ô∏è‚É£ **Point Mort**
- Le **nombre de jours** pour atteindre le SR dans l'ann√©e
- Si SR = jour 100, vous √™tes rentable √† partir du 101e jour
- **Montre la rapidit√©** pour atteindre l'√©quilibre
